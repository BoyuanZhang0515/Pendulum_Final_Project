```python
# Import Packages
import math
import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction
```

```python
# Basic settings
r = 0.1
L = 1
g = 9.81
theta_0 = 1/8* math.pi

tau = 2*math.pi * (L/g)**0.5
tf = 40
#td = tf/tau
k = 10
A = math.pi*r**2
m = 1
```

```python
# Figure 1: The Simple sphere Pendulum looping different dt
lines_3 = [];
Lx = 20;# in
Ly = 10;# in
plt.figure(figsize=[Lx, Ly])
ax = plt.axes([0.15, 0.15, 0.8, 0.8], xlim=(0, 10), ylim=(-1, 2))

plt.title("The Simple Pendulum Graphic (Looping different dt with $\Theta_0$= $\pi/8$)",fontsize = 16)
plt.xlabel("The Dimenionless time (t/$tau$)",fontsize = 16)
plt.ylabel("$\Theta$(Radians)",fontsize = 16)
dt_3 = [0.0001,0.001,0.01]
#ax = plt.axes([0.15, 0.15, 0.8, 0.8], xlim=(0, 5), ylim=(-0.8, 0.8))
for dt_tmp in dt_3:
    
    tM = math.floor(tf/(dt_tmp))+1
    t=np.linspace(0,tf,tM);
    
    Theta = np.zeros(tM)
    Theta_dot = np.zeros(tM)
    Theta[0] = theta_0
    Theta_dot[0] = 0
    for j in range(tM-1):
        Theta_dot_dot_tmp = -g/L*math.sin(Theta[j])
        Theta[j+1] = Theta_dot[j]*dt_tmp+Theta[j]
        Theta_dot[j+1] = Theta_dot[j] + Theta_dot_dot_tmp * dt_tmp
    lines_3 += plt.plot(t[1:]/tau,Theta[1:], "-", label="$" + str(round(dt_tmp,4)) + "$");
        
labels_3 = [l.get_label()+'s' for l in lines_3]
plt.legend(lines_3, labels_3, frameon=False, title="$\Delta t\;$",fontsize = 16)
plt.show()
```

```python
# Figure 2: The Simple Pendulum looping different Theta 0
Lx = 20;# in
Ly = 10;# in
plt.figure(figsize=[Lx, Ly])
ax = plt.axes([0.15, 0.15, 0.8, 0.8], xlim=(0, 10), ylim=(-1, 2))
plt.title("The Simple Pendulum Graphic (Looping different $\Theta_0$ with dt = 0.0001s)",fontsize = 20)
plt.xlabel("The Dimenionless time (t/$\tau$)",fontsize = 20)
plt.ylabel("$\Theta$(Radians,fontsize =20)",fontsize = 20)


Theta_0s = [1/4,1/8,1/16,1/32]
lines_4 = []
for theta_0_tmp in Theta_0s:
    dt_tmp = 0.0001
    tM = math.floor(tf/(dt_tmp))+1
    t=np.linspace(0,tf,tM);
    theta_0_tmp_pi = theta_0_tmp*math.pi
    
    Theta = np.zeros(tM)
    Theta_dot = np.zeros(tM)
    Theta[0] = theta_0_tmp_pi
    Theta_dot[0] = 0
    for j in range(tM-1):
        Theta_dot_dot_tmp = -g/L*math.sin(Theta[j])
        Theta[j+1] = Theta_dot[j]*dt_tmp+Theta[j]
        Theta_dot[j+1] = Theta_dot[j] + Theta_dot_dot_tmp * dt_tmp
    lines_4 += plt.plot(t/tau,Theta, "-", label="$\pi$ / "+str(int(1/theta_0_tmp)));
        
labels_4 = [l.get_label() for l in lines_4]
plt.legend(lines_4, labels_4, frameon=False, title="$\Theta_0$ (Degrees)",fontsize =20)
plt.show()
```

```python
# Figure 3: A Damping Pendulum Graphic (Looping different drag force constants k)
Lx = 20;# in
Ly = 10;# in
plt.figure(figsize=[Lx, Ly])
ax = plt.axes([0.15, 0.15, 0.8, 0.8], xlim=(0, 20), ylim=(-0.4, 0.4))
plt.title("A Damping Pendulum Graphic (Looping different drag force constants k with dt = 0.0001s,$\Theta_0=\pi/8$",fontsize = 20)
plt.xlabel("The Dimenionless time (t/$tau$)",fontsize =20)
plt.ylabel("$\Theta$(Radians)",fontsize =20)

#ax= plt.axes([0.15, 0.15, 0.8, 0.8], xlim=(0, 100), ylim=(-1, 2))


ks = [1,5,10]
lines_4_2 = []
for k in ks:
    dt_tmp = 0.0001
    tM = math.floor(tf/(dt_tmp))+1
    t=np.linspace(0,tf,tM);
    
    Theta_0 = (1/8)*math.pi
    Theta = np.zeros(tM)
    Theta_dot = np.zeros(tM)
    Theta[0] = Theta_0
    Theta_dot[0] = 0
    for j in range(tM-1):
        Theta_dot_dot_tmp = -k*A/m*Theta_dot[j]-g/L*math.sin(Theta[j])
        Theta[j+1] = Theta_dot[j]*dt_tmp+Theta[j]
        Theta_dot[j+1] = Theta_dot[j] + Theta_dot_dot_tmp * dt_tmp
    lines_4_2 += plt.plot(t/tau,Theta, "-", label= str(k));
        
labels_4_2 = [l.get_label() for l in lines_4_2]
plt.legend(lines_4_2, labels_4_2, frameon=False, title="Drag Force Constant k ($kgm^{-2}s^{-1}$) ",fontsize =20)
plt.show()
```

```python
# The Rod Pendulum Graphic (Looping different k)
a = 0.1
m = 1
Lx = 20;# in
Ly = 10;# in
plt.figure(figsize=[Lx, Ly])
ax = plt.axes([0.15, 0.15, 0.8, 0.8], xlim=(0, 20), ylim=(-0.4, 0.4))
plt.title("The Rod Pendulum Graphic (Looping different k with dt = 0.0001s, $\Theta_0=\pi/8$ ",fontsize =20)
plt.xlabel("The Dimenionless time (t/$\tau$)",fontsize =20)
plt.ylabel("$\Theta$(Radians)",fontsize =20)

#ax= plt.axes([0.15, 0.15, 0.8, 0.8], xlim=(0, 100), ylim=(-1, 2))


ks = [1,5,10]
lines_5_1 = []
for k in ks:
    dt_tmp = 0.0001
    tM = math.floor(tf/(dt_tmp))+1
    t=np.linspace(0,tf,tM);
    theta_0 = 1/8*math.pi
    
    Theta = np.zeros(tM)
    Theta_dot = np.zeros(tM)
    Theta[0] = theta_0
    Theta_dot[0] = 0
    for j in range(tM-1):
        Theta_dot_dot_tmp = -3/2*(g/L)*math.sin(Theta[j])-k*2*a*(L/m)*Theta_dot[j]
        Theta[j+1] = Theta_dot[j]*dt_tmp+Theta[j]
        Theta_dot[j+1] = Theta_dot[j] + Theta_dot_dot_tmp * dt_tmp
    lines_5_1 += plt.plot(t/tau,Theta, "-", label=str(k));
        
labels_5_1 = [l.get_label() for l in lines_5_1]
plt.legend(lines_5_1, labels_5_1, frameon=False, title="Drag Force Constant k ($kgm^{-2}s^{-1}$)",fontsize =20)
plt.show()
```
